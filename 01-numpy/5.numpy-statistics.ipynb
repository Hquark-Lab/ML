## 常见的统计运算

```python
import numpy as np
```

```python
x = np.arange(5000)
```

```python
np.max(x)
```

```python
X = x.reshape(50,-1)
X.shape
```

```python
np.min(X,axis=1) #每行最小值
```

```python
np.max(X,axis=0)#每列最大值
```

```python
np.sum(X,axis=1)
```

```python
np.median(X,axis=0)#取中位数
```

```python
np.std(X) #标准差
```

```python
np.var(X) #方差
```

```python
Z = np.random.normal(0,10,4000)
np.mean(Z), np.std(Z)
```

```python
np.ptp(Z) #元素最大值和最小值的差
```

### 百分位数

```python
q = np.linspace(0,100,4)
q,np.percentile(X,q)
```
