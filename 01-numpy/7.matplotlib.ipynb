```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-10,10,100)
```

```python
y = x * x
```

```python
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.plot(x,y1,color = 'red',linestyle=':',label='sin(x)')
plt.plot(x,y2,label='cos(x)')
plt.ylim(-2,2)
plt.ylabel("y label")
plt.legend() #
plt.show()
```
