## 数组的合并

```python
X = np.arange(12).reshape(3,4)
```

```python
X
```

```python
a = np.array([1,1,1,1])
```

```python
np.vstack((X,a)) #竖直方向合并
```

```python
C = np.full((3,2),34)
np.hstack((X,C))
```
