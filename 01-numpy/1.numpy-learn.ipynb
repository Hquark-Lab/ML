## 随机数

```python
np.random.random() #[0, 1)
```

```python
np.random.random((3,6))
```

```python
np.random.rand(3,4)#这里不需要传入元组
```

```python
np.random.randint(5) #[0, 5)
```

```python
np.random.randint(5,10,size=8) #[5, 10)生成8个随机数
```

```python
np.random.randint(5,10,size=(3,4))
```

## 标准正态分布

```python
np.random.randn(4)#生成来自标准正态分布（均值为0，标准差为1）的随机数(4) 表示生成一个包含 4 个元素的一维数组
```

## 高斯分布数据

```python
np.random.normal(10,200,5)#loc = 10：均值（分布的中心位置）
#scale = 200：标准差（分布的离散程度）
#size = 5：输出数组的形状（生成5个随机数）
```

## 均匀分布随机数的函数

```python
np.random.uniform(1,5)#low = 1：区间下限（包含）；high = 5：区间上限（不包含）；size：未指定时默认生成 1 个随机数
```
